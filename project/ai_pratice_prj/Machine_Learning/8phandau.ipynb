## 8.3 Recursive Feature Elimination

```python
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# c d liu
filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(filename, names=names)

# Chuyn tt c cc ct thnh s v loi b cc gi tr khng hp l
dataframe = dataframe.apply(pd.to_numeric, errors='coerce')
dataframe = dataframe.dropna()

# To mng d liu
array = dataframe.values
x = array[:, 0:8]  # Cc tnh nng
y = array[:, 8]    # Ct 'class'

# Chun ha d liu
scaler = StandardScaler()
x = scaler.fit_transform(x)

# Khi to Logistic Regression vi tng s vng lp v thut ton 'lbfgs'
model = LogisticRegression(max_iter=200)  # Tng s vng lp ln 200

# p dng RFE  chn 3 tnh nng
rfe = RFE(model, n_features_to_select=3)
fit = rfe.fit(x, y)

# In kt qu
print('Num Features: %d' % fit.n_features_)
print('Selected Features: %s' % fit.support_)
print('Feature Ranking: %s' % fit.ranking_)
```
